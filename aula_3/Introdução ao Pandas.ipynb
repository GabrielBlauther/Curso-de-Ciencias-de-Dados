### Pegar 1 coluna (e os pd.Series)

```python
#Serie é uma coluna do dataframe
#São dois colchetes pois está pegando mais de uma coluna do dataframe
produtos = vendas_df[['Produto', 'ID Loja', 'Valor Unitário']]
display(produtos)
```

### Método .loc para pegar partes do Dataframe [linhas,colunas]

```python
# pegar uma única linha
display(vendas_df.loc[10])
```

```python
# pegar os dados da linha 5 a linha 50 linha coluna
display(vendas_df.loc[5:50])
```

```python
# pegar linhas que correspondem a uma condição
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping_df)
```

```python
# pegar várias linhas e colunas usando o loc
#loc[linhas,colunas]
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', [ "Produto", "Quantidade", "Valor Unitário"]]
display(vendas_norteshopping_df)
```

```python
# pegar 1 valor específico
print(vendas_df.loc[30,'Produto'])
```

### Adicionar colunas no Dataframe

```python
# a partir de uma coluna que existe
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.02
display(vendas_df)
```

```python
# criar uma coluna com valor padrão
#.loc[linhas,colunas]
#vendas_df['Imposto'] = 0
vendas_df.loc[:, "Imposto"] = 0
display(vendas_df)
```

### Adicionar linhas em Dataframe

```python
#Lendo outro arquivos de vendas para um novo dataframe
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
display(vendas_dez_df)
```

```python
vendas_df
```

```python
#Adiciona o Dataframe vendas_dez_df no final do dataframe vendas_df
#Método append sempre adiciona
# Concatenar DataFrames
vendas_df_2 = pd.concat([vendas_df, vendas_dez_df], ignore_index=True)
display(vendas_df_2)
```

### Excluir linhas e colunas

```python
#Excluir a linha 0
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

### Tratar Valores Vazios em Dataframe

```python
display(vendas_df_2)
```

```python
# preencher valores vazios
# preencher com a média da coluna Comissão (fillna), arredondando os valores para 2 casas decimais
#vendas_df_ap['Comissão'] = vendas_df_ap['Comissão'].fillna(0)

vendas_df_2['Comissão'] = np.round(vendas_df_2['Comissão'].fillna(vendas_df_2['Comissão'].mean()),2)
display(vendas_df_2)
```

```python
#Método append sempre adiciona
vendas_df_3 = vendas_df.append(vendas_dez_df)
display(vendas_df_3)
```

```python
# preencher com o último valor (ffill)
vendas_df_3 = vendas_df_3.ffill()
display(vendas_df_3)
```

```python
#Resetar os indices do Dataframe
vendas_df_2.reset_index(inplace = True, drop = True)
display(vendas_df_2)
```
