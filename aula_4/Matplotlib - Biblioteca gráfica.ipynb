```python
import numpy as np
import pandas as pd
```

```python
# O matplotlib.pyplot é uma coleção de funções e estilos que fazem com que o Matplotlib funcione como o Matlab.
import matplotlib.pyplot as plt
```

```python
linha1 = np.array([1,2,3,4])
linha2 = [1,3,9,16]
```

```python
plt.plot(linha1)
plt.plot(linha1,linha2)
#Range dos eixos [xmin,xmax,ymin,ymax]
plt.axis([1,4,1,16])
plt.xlabel('Variável 1')
plt.ylabel('Variável 2')
plt.title('Teste Plot')
plt.show()
#linha1 = np.array([1,2,3,4])
#linha2 = [1,3,9,16]
```

```python
#Gera um numpy array de 0 a 5 (excludente) com intervalos de 0.2
t=np.arange(0.,5.2,0.2)
print (t)
#Gera um gráfico de linhas com diferentes marcações
#eixo x, eixo y, cor,marcador
plt.plot (t,t,'r--',
          t,t**2,'bs',
          t,t**3,'m+')
plt.show()
```

### Scatterplot (Gráfico de dispersão)

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,5,6,8,4,8]
x1 = [5,2,4,5,6,8,4,8]
y1 = [1,1,3,4,5,6,7,10]
plt.scatter(x, y, label = 'Pontos', color = 'r', marker = 'o', s = 75)
plt.scatter(x1, y1, label = 'Quadrados', color = 'b', marker = 's', s = 50)

plt.xlabel('Variável x')
plt.ylabel('Variável y')
plt.title('Teste Scatter Plot')
plt.legend()
plt.show()
```

### Visualização para séries temporais

```python
# Cria um dataframe de serie temporal
ts = pd.Series(np.random.randn(500)*10, index=pd.date_range('1/1/2021', periods = 500))
display(ts)
```

```python
#Retorna o mesmo dataframe com somas acumulativas
ts = ts.cumsum()
ts.plot()
```

```python
# Cria um dataframe 500x4 com valores randomicos, contendo as colunas A,B,C e D e o mesmo indice do dataframe ts
df = pd.DataFrame(np.random.randn(500, 4), index = ts.index, columns = ['A', 'B', 'C', 'D'])
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.cumsum()
plt.figure()
df.plot()
#loc: localização da legenda
plt.legend(loc = 'best')
plt.xlabel('Tempo - dia')
plt.ylabel('Variação')
plt.title('Dados Acumulados ao longo do tempo - Time Series')
```

```python
# Plot e Scatter
fig = plt.figure()
#fig.add_subplot(nrows, ncols, index): 1 linha, 2 colunas e gráfico da esquerda
ax1 = fig.add_subplot(1,2,1)
ax1.plot(np.random.randn(50), color='red')
#fig.add_subplot(nrows, ncols, index): 1 linha, 2 colunas e gráfico da direita
ax2 = fig.add_subplot(1,2,2)
ax2.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```python
# Plots diversos
_, ax = plt.subplots(2,3)

ax[0,1].plot(np.random.randn(50), color = 'green', linestyle = '-')
ax[0,2].hist(np.random.randn(100))
ax[1,0].hist(np.random.randn(50))
ax[1,2].scatter(np.arange(50), np.random.randn(50), color = 'red')
plt.show()
```
